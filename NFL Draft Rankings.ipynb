## SportingNews

```python
sportingnews = 'https://www.sportingnews.com/us/nfl/news/nfl-draft-grades-2020-best-worst-classes/2bs9w101joku1lruu8r1xwt9f'
r = requests.get(sportingnews)
soupSporting = BeautifulSoup(r.content, 'html.parser')
print(soupSporting.prettify())
```

```python
soupSporting.find_all('ul')[42].get_text()
```

```python
teams = []
grade = []
for i in range(2,34):
    teams.append(soupSporting.find_all('h3')[i].get_text())
for j in range(11,43):
    grade.append(soupSporting.find_all('ul')[j].get_text())
```

## DraftWire

<font size = 3>This was the most difficult website to scrape. Each team has a unique URL. For the Arizona Cardinals, the url was not as easily scraped, so I did it manually. Raiders and 49ers both had team names that were not 'h2' on their respective websites, so those were also manually input.

```python
d_url = 'https://draftwire.usatoday.com/2020/04/26/2020-nfl-draft-grades-full-every-team/'
r_d = requests.get(d_url)
d_soup = BeautifulSoup(r_d.content, 'html.parser')

d = []
for i in range(2):
    d.append(d_soup.find_all('h2')[i].get_text())
d
```

```python
draftwire_url = []
draft_team = []
for i in range(2, 33):
    draftwire_url.append(('https://draftwire.usatoday.com/2020/04/26/2020-nfl-draft-grades-full-every-team/'+str(i)+'/'))
    
for j in draftwire_url:
    r_draftwire = requests.get(j)
    draft_soup = BeautifulSoup(r_draftwire.content, 'html.parser')
    
    for k in range(len(draft_soup.find_all('h2'))):
        draft_team.append(draft_soup.find_all('h2')[k].get_text())

#raiders was not a h3, so I just manually inserted it
#same with 49ers
draft_team.insert(30, 'Las Vegas Raiders')
draft_team.insert(52, 'San Francisco 49ers')
draft_team =  d + draft_team

draftwire_teams = draft_team[::2]
draftwire_grades = draft_team[1::2]
```

```python
draftwire = pd.DataFrame(draftwire_teams, columns = ['Team'])
draftwire['DRAFTWIRE (Luke Easterling)'] = draftwire_grades

draftwire['DRAFTWIRE (Luke Easterling)'] = draftwire['DRAFTWIRE (Luke Easterling)'].str.split(' ', expand = True)[1]
```

```python
df8 = pd.merge(df7, draftwire)
df8.head()
```

## Colored Table

<font size = 3>Chunk of code below specifies the formatting that was implemented on the styled table.

```python
def rowcolor(s, threshold, column):
    is_max = pd.Series(data=False, index=s.index)
    is_max[column] = s.loc[column] == threshold
    return ['background-color: thistle' if is_max.any() else '' for v in is_max]


aa = Team_GPA_DF.reset_index(drop = True).style.background_gradient(subset = ['Average Team GPA'], cmap = 'RdYlGn').set_caption('NFL Draft Grades From Popular Football Sites').set_properties(**{'text-align':'center',
                                                                                                                    'border-color': 'black','border-style' :'solid' ,
                                                                                                                        'border-width': '1px'}).apply(rowcolor, 
                                                                                                                                                                 threshold = 'Website Average GPA', column = ['Team']
                                                                                                                                                                        , axis = 1).set_precision(2).set_table_styles([dict(selector = 'th',
                                                                                                                                                                                props = [('max-width', '150px')])]).set_table_styles([{'selector':'caption',
                                                                                                                                                                                                                        'props': [('color','black'),('font-size', '32px')]}]).hide_index()
```

```python
aa
```

```python
html = aa.render()
```
